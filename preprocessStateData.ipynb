```python
import pandas as pd
import json


DATA_DIR = "data/StateAugmentationData"

FILE_PATHS = [
    'median_income.json',
    'poverty_rates.json',
    'state_abbreviations.json',
    'state_by_region.json',
    'state_gini.json',
    'state_gun_law_strength_ranking.json',
    'unemployment_rate.json'
]
```

```python
data = {}
for file_path in FILE_PATHS:
    with open(f"{DATA_DIR}/{file_path}") as f:
        data[file_path] = json.load(f)

# reverse state_abbreviations.json
data['state_abbreviations.json'] = {v: k for k, v in data['state_abbreviations.json'].items()}

states = set(state for file_path in FILE_PATHS for state in data[file_path].keys())

states = sorted(list(states))

assert len(states) == 51, f"expected 51 states, got {len(states)}"

states = data[FILE_PATHS[0]].keys()
df = pd.DataFrame({'state': list(states)})
for file_path in FILE_PATHS:
    df[file_path] = [data[file_path].get(state) for state in states]

df = df.rename(columns={
    'median_income.json': 'median_income',
    'poverty_rates.json': 'poverty_rate',
    'state_abbreviations.json': 'state_abbr',
    'state_by_region.json': 'region',
    'state_gini.json': 'gini',
    'state_gun_law_strength_ranking.json': 'gun_law_strength_ranking',
    'unemployment_rate.json': 'unemployment_rate'
})

df.to_csv('data/state_data.csv', index=False)

df.head()
```
